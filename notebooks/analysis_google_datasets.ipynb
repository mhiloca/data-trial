```python
import pandas as pd
import psycopg2 as pg
```

### Create engine and dataframes from uploaded tables
- company_profiles_google_maps
- customer_reviews_google

```python
engine = pg.connect("dbname='clever' user='clever' host='postgres_clever' port='5432' password='clever'")

company_profiles_df = pd.read_sql('select * from public.company_profiles_google_maps', con=engine)
customer_reviews_df = pd.read_sql('select * from public.customer_reviews_google', con=engine)
```

### Check on columns and real estate categories

```python
company_profiles_df.columns
```

```python
company_profiles_df[company_profiles_df.category.str.startswith('Real estate')].category.unique()
```

### Analyse customer reviews per real estate company
- Filter `company_profiles_df` by real estate companies related only
- Join with `customer_reviews_df` to grab rating given to each review

```python
real_estate_categories = ['Real estate agent', 'Real estate agency', 'Real estate consultant', 'Real estate']
real_estate_df = company_profiles_df[company_profiles_df.category.isin(real_estate_categories)]
real_estate_reviews_df = real_estate_df.merge(customer_reviews_df, on=['place_id', 'google_id', 'rating'] , how='inner', sort=True, suffixes=['_profile', '_customer_review'])
real_estate_reviews_df.head()
```

### Filter real estate companies in the city of Miami

```python
miami_real_estate_reviews_df = real_estate_reviews_df[(real_estate_reviews_df.city == 'Miami') & (real_estate_reviews_df.state.isin(['Florida', 'FL']))] 
miami_real_estate_reviews_df.head()
```

## The 10 real estate companies with more 5 ratings in Miami, FL

```python
miami_real_estate_reviews_df.loc[miami_real_estate_reviews_df.review_rating == 5, 'num_five_rating'] = 1
(
    miami_real_estate_reviews_df[['place_id', 'name', 'rating', 'num_five_rating']]
    .groupby(['place_id', 'name', 'rating'])
    .count().sort_values(by='num_five_rating', ascending=False).head(10)
)
```
